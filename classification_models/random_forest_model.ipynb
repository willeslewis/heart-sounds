```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import LabelEncoder
```

```python
train=pd.read_csv('/Users/William/Desktop/heart-sounds/preprocessed_data/training.csv')
```

```python
train.head()
```

```python
train = train.sample(frac=1)
Xtrain = train.drop('type',axis=1)
ytrain = train['type']
lbl = LabelEncoder()
lbl.fit(ytrain)
ytrain = lbl.transform(ytrain)
```

```python
rf = RandomForestClassifier()

gsrf = GridSearchCV(rf, param_grid = {'n_estimators':[10,50,150,250]}, n_jobs = -1,return_train_score=True,cv=10)

gsrf.fit(Xtrain,ytrain)
```

```python
gsrf.get_params().keys()
```

```python
gsrf.cv_results_
```

```python
test=pd.read_csv('/Users/William/Desktop/heart-sounds/preprocessed_data/test.csv')
Xtest=test.drop('type',axis=1)
ytest=test['type']
ytestnum=lbl.transform(ytest)
```
